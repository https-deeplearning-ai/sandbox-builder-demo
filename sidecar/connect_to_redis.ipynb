# Connect to Redis

This notebook demonstrates connectivity to Redis from Jupyter Lab.

**Note:** Inside Docker Compose, we connect using the service name `redis` as the hostname (not `localhost`). The Redis service is exposed on `localhost:6379` from the host machine.

```python
# Install redis-py if not already installed
!pip install redis -q
```

```python
import redis

# Connect to Redis (using service name in Docker Compose network)
r = redis.Redis(host='localhost', port=6379, decode_responses=True)  # use host=redis when docker compose

# Test connection
print(f"Ping: {r.ping()}")
```

```python
# Basic Redis operations

# Set a key
r.set('greeting', 'Hello from Jupyter!')

# Get the key
value = r.get('greeting')
print(f"Value: {value}")
```

```python
# List all keys
keys = r.keys('*')
print(f"All keys: {keys}")
```
