## 1. A simple call with the OpenAI API

```python
from openai import OpenAI

# Instantiate the client
openai_client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY", "123"),
    base_url=os.getenv("OPENAI_BASE_URL", "http://jupyter-api-proxy.internal.dlai/rev-proxy/anthropic"),
)
```

```python
completion = openai_client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Make up a user."},
    ],
)

print(completion.choices[0].message.content)
```

## 2. A simple call with the Anthropic API

```python
import anthropic
import os

anthropic_client = anthropic.Anthropic(
    api_key=os.getenv("ANTHROPIC_API_KEY", "sk-xxx",
    base_url=os.getenv("ANTHROPIC_BASE_URL", "http://jupyter-api-proxy.stg.internal.dlai/rev-proxy/anthropic"),  # staging: http://jupyter-api-proxy.stg.internal.dlai/rev-proxy/anthropic
)
```

```python
message = anthropic_client.messages.create(
    model="claude-opus-4-8",
    max_tokens=1024,
    system="You are a helpful assistant.",
    messages=[
        {"role": "user", "content": "Make up a user?"},
    ],
)
# response.content is a list of content blocks; print the text
for block in message.content:
    if block.type == "text":
        print(block.text)
```
